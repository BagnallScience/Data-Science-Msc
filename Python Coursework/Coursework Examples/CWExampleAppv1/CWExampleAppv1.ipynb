```python
# all function defs 

def display_line(symbol="*",line_size=10):
    line=line_size*symbol
    print(line)

def display_msg(msg="hello"):
    display_line()
    print("!!!"+msg+"!!!")
    display_line()

def display_menu():
    display_msg("Menu")
    print("h) Say 'hi'")
    print("b) Say 'bye'")
    print("d) Display Marks")
    print("a) Add a new mark")
    print("s) Save Student Marks")
    print("l) Load Student Marks")
    print("e) Exit")

# menu actions     
def say_hi():
    print("hi")
    
def say_bye():
    print("bye")
    
def add_mark():
    print("add a new student mark")
    studentName=input("please enter student name")
    studentMark=int(input("please enter student mark"))
    marks.append([studentName,studentMark])
    display_marks()
    
def display_marks():
    print("Student Marks")
    for mark in marks:
        print(mark[0],mark[1])
#####
# File Operations Functions 
#####
def load_marks():
    StudentMarks=[]
    
    f=open("marks.txt","r")
    for mark in f:
        clean_mark=mark.strip()
        studentMark=clean_mark.split(",")
        
        StudentMarks.append(studentMark)
    print("student Marks\n",StudentMarks)
    
    f.close()
    return StudentMarks

    
def save_marks(studentMarks):
    print("saving students' marks to the file")
    f=open("marks.txt","w")
    for mark in studentMarks:
        record=mark[0]+","+str(mark[1])+"\n"
        f.write(mark[0]+","+str(mark[1])+"\n")
    f.close()
```

```python
############
#----Main--- 
############
display_menu()
marks=[]

while True:
    opt=input("please enter your option")
    if opt=="h":
        say_hi()
    elif opt=="b":
        say_bye()
    elif opt=="d":
        display_marks()
    elif opt=="a":
        add_mark()
    elif opt=="l":
        marks=load_marks()
    elif opt=="s":
        save_marks(marks)
    elif opt=="e":
        print("exit")
        break
    else:
        print("sorry")

```
