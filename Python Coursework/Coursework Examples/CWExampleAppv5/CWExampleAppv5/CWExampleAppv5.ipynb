```python
# all function defs 

def display_line(symbol="*",line_size=10):
    line=line_size*symbol
    print(line)

def display_msg(msg="hello"):
    display_line()
    print("!!!"+msg+"!!!")
    display_line()

    
def display_menu():
    display_msg("Menu")
    print("d) Display Marks")
    print("a) Add a new mark")
    print("s) Save Student Marks")
    print("l) Load Student Marks")
    print("p) Plot Marks")
    print("c) Clear Screen")
    print("t) Transfer to DB" )
    print("r) retrieve from DB")
    print("e) Exit")

# menu actions     

def plot_marks(): # new 
    dfMarks.plot(x="name",y="mark",kind="bar",title="Student Marks")
    plt.show()
    
def add_mark():
    print("add a new student mark")
    studentName=input("please enter student name")
    studentMark=int(input("please enter student mark"))
    dfMarks.loc[len(dfMarks)]=[studentName,studentMark]
    display_marks()
    
def display_marks():
    print("Student Marks")
    display(dfMarks)
```

```python
############
#----Main--- 
############
import fileOperations as fo

display_menu()

dfMarks=fo.load_marks()

while True:
    opt=input("please enter your option")
    if opt=="l":
        dfMarks=fo.load_marks()
    elif opt=="d":
        display_marks()
    elif opt=="a":
        add_mark()
    elif opt=="s":
        fo.save_marks(dfMarks)
    elif opt=="r":
        dfMarks=fo.load_from_db()
    elif opt=="t":
        fo.save_to_db(dfMarks)     
    elif opt=="p":
        plot_marks()
    elif opt=="c":
        clear_output()
        display_menu()
    elif opt=="e":
        print("exit")
        break
    else:
        print("sorry")

```
